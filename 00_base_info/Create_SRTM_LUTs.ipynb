```python
import geopandas
import math
from tqdm.notebook import tqdm
import rsgislib.vectorattrs
```

```python
base_tiles = "00_base_info/gmw_degree_tiles_fnl.geojson"
srtm_tiles = "04_fill_gaps/imgs_lut.gpkg"
srtm_tiles_pop = "04_fill_gaps/imgs_lut_pop.geojson"
```

```python
rsgislib.vectorattrs.add_geom_bbox_cols(srtm_tiles, "30_100cm", srtm_tiles_pop, "imgs_lut_pop", out_format = 'GeoJSON', min_x_col = 'MinX', max_x_col = 'MaxX', min_y_col = 'MinY', max_y_col = 'MaxY')
```

```python
base_gpd_df = geopandas.read_file(srtm_tiles_pop)
join_gpg_df = geopandas.read_file(base_tiles)

```

```python
files_names = base_gpd_df["filename"].values
out_tiles_names = list()
for filename in files_names:
    tilename = filename.split("_")[0]
    out_tiles_names.append(tilename)
```

```python
base_gpd_df["tilenames"]=out_tiles_names
```

```python
base_gpd_df
```

```python
for col in ["MinX", "MinY"]:
    out_col = list()
    for val in base_gpd_df[col].values:
        out_col.append(math.floor(val))
    base_gpd_df[col] = out_col
    
for col in ["MaxX", "MaxY"]:
    out_col = list()
    for val in base_gpd_df[col].values:
        out_col.append(math.ceil(val))
    base_gpd_df[col] = out_col
                       
```

```python
join_gpg_df
```

```python
base_MinX = base_gpd_df["MinX"].values
base_MinY = base_gpd_df["MinY"].values
base_MaxX = base_gpd_df["MaxX"].values
base_MaxY = base_gpd_df["MaxY"].values

join_MinX = join_gpg_df["MinX"].values
join_MinY = join_gpg_df["MinY"].values
join_MaxX = join_gpg_df["MaxX"].values
join_MaxY = join_gpg_df["MaxY"].values
join_prj_name = join_gpg_df["prj_name"].values
join_tile_name = join_gpg_df["tile_name"].values

```

```python
out_prj_name = list()
out_tile_name = list()

for b_minX, b_minY, b_maxX, b_maxY in zip(base_MinX, base_MinY, base_MaxX, base_MaxY):
    first = True
    
    bc_x = b_minX + (b_maxX - b_minX)/2
    bc_y = b_minY + (b_maxY - b_minY)/2
    #print("{} , {}".format(bc_x, bc_y))
    min_dist = 0.0
    min_prj = ""
    min_tile = ""
    for j_minX, j_minY, j_maxX, j_maxY, j_prj, j_tile in zip(join_MinX, join_MinY, join_MaxX, join_MaxY, join_prj_name, join_tile_name):
        jc_x = j_minX + (j_maxX - j_minX)/2
        jc_y = j_minY + (j_maxY - j_minY)/2
        
        dist = math.sqrt((bc_x - jc_x)**2 + (bc_y - jc_y)**2)
        
        #print("\t{} , {} = {}".format(jc_x, jc_y, dist))
        
        if first and (dist < 2):
            min_dist = dist
            min_prj = j_prj
            min_tile = j_tile
            first = False
            
            #print("\t{} , {}".format(bc_x, bc_y))
            #print("\t{} , {}".format(jc_x, jc_y))
            #print("\tDist: {}\n".format(dist))
            
        elif dist < min_dist:
            min_dist = dist
            min_prj = j_prj
            min_tile = j_tile
            
            #print("\t{} , {}".format(bc_x, bc_y))
            #print("\t{} , {}".format(jc_x, jc_y))
            #print("\tDist: {}\n".format(dist))
            
    if not first:
        out_prj_name.append(min_prj)
        out_tile_name.append(min_tile)
    else:
        out_prj_name.append("")
        out_tile_name.append("")

        
```

```python
base_gpd_df["prj_name"] = out_prj_name
base_gpd_df["join_tile_name"] = out_tile_name
```

```python
base_gpd_df
```

```python
srtm_prj_tiles_lut = dict()
srtm_tiles_prj_lut = dict()

for tile, prj in zip(base_gpd_df["tilenames"].values, base_gpd_df["prj_name"].values):
    if prj not in srtm_prj_tiles_lut:
        srtm_prj_tiles_lut[prj] = list()
    
    if tile not in srtm_prj_tiles_lut[prj]:
        srtm_prj_tiles_lut[prj].append(tile)
    
    if tile not in srtm_tiles_prj_lut:
        srtm_tiles_prj_lut[tile] = prj
```

```python
import rsgislib.tools.utils
```

```python
rsgislib.tools.utils.write_dict_to_json(srtm_prj_tiles_lut, "00_base_info/gmw_srtm_projects_luts.json")
rsgislib.tools.utils.write_dict_to_json(srtm_tiles_prj_lut, "00_base_info/gmw_srtm_tiles_luts.json")
```

```python
base_gpd_df = base_gpd_df.drop(columns=["filename", "path"])

```
